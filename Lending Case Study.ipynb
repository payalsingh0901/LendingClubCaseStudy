```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

#dataframe df
#file_path is the Location of your Loan source file

file_path = 'Loan1.csv'
df = pd.read_csv(file_path)

#Creating another column loan_status_binary to determine if the loan is defaulted or not and add it to the dataframe
# Value of 1 for defaulted and 0 for the other status (Current and Fully Paid)
df['loan_status_binary'] = df['loan_status'].apply(lambda x: 1 if x == 'Charged Off' else 0)

#1: Draw a barplot with grade and loan_status_binary and see if there is a relationship
sns.barplot(x='grade', y='loan_status_binary', data=df, order=sorted(df['grade'].unique()))
plt.title('Relationship between Grades and Default Rates')
plt.xlabel('Grade')
plt.ylabel('Loan Default Rate')
plt.show()
#Conclusion:This shows that as the grades go from A to G, number of loan default rates increases which shows us a pattern based on univariate

#2: To check if there is any relationship between Term and Loan Default status
sns.barplot(x='term', y='loan_status_binary', data=df, palette='viridis')
plt.title('Comparison: Loan Term vs. Default Status')
plt.xlabel('Loan Term')
plt.ylabel('Default Rate')
plt.show()
#Conclusion:This shows that as the Loan term increases, loan default also increases and is directly propertional

#3: Verified income vs loan default status
sns.barplot(x='verification_status', y='loan_status_binary', data=df, palette='viridis')
plt.title('Comparison: Verification Status vs. Default Status')
plt.xlabel('\nVerification Status')
plt.ylabel('Default Rate')
plt.show()
#Conclusion:This shows that if the income is verified, there are less chances of getting loan defaulted compared to those where it isnt verified
```
